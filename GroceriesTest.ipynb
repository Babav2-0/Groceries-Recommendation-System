```python
#Helping Functions to calculate cosine
def square(col):
    sum=0
    for i in col:
        sum+=i*i
    return sum
def product(a,b):
    p=a*b
    return sum(p)
```

```python
recommended=dict()
for i in col:
    recommended[i]=set()
```

### Slope one Algorithm
- Product to product recommendation.
- Use collaborative filtering technique.
- Calculate similarity of products between users

```python
si=[]
for i in range(len(col)):
    for j in range(len(col)):
        if i!=j:
            similarity_index=product(data[col[i]],data[col[j]])
            similarity_index/=((sum(data[col[i]])*sum(data[col[j]]))**0.5)
            #si.append(index)
            if(similarity_index>=0.1):
               recommended[col[i]].add(col[j])
```
